```python
full_df=pd.read_csv('Scraped_Earbud_Featured_Page.csv')
```

```python
full_df['Product_ID']=full_df['URL'].apply(lambda st: st[st.find("dp/")+3:st.find("/ref")])
full_df['reviewmeta_rating']=''
full_df['decision']=''
```

```python
full_df
```

```python
driver = webdriver.Chrome("C:\chromedriver_win32\chromedriver.exe")
clicked_agree=0
for index,row in full_df.iterrows():
    #print(row['Product_ID'])
    print(index)
    reviewmeta_url='https://reviewmeta.com/amazon/'+row['Product_ID']
    print(reviewmeta_url)
    driver.get(reviewmeta_url)
    
    #agree_button=driver.find_element_by_id('terms_accepted')
    #print(agree_button)
    if(clicked_agree==0):
        driver.find_element_by_id('terms_accepted').click()
        clicked_agree=1
    #meta_rating=driver.find_element_by_id('adjusted-rating-large')
    #print(meta_rating)

    #time.sleep(10)
    ratings=driver.find_elements_by_xpath("//label[contains(@class,'orig-rating')]")
    print(len(ratings))
    print(ratings[1].text)
    reviewmeta_rating=ratings[1].text
    if reviewmeta_rating== "N/A":
        print("Looks fake")
        #reviewmeta_rating="N/A"

        full_df.at[index,'decision']='Fake'
    else:
        print("Looks good")

        full_df.at[index,'decision']='Good'
        reviewmeta_rating=reviewmeta_rating.split("/")[0]

    full_df.at[index,'reviewmeta_rating']=reviewmeta_rating    
        
#    except:
#        print('----Needs reviwing---- ')
#        print(reviewmeta_url)
    
    
        
    
    
    
```
